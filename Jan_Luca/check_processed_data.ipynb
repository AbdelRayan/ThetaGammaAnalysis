```python
import h5py
import os
import numpy as np
```

```python
#the ratinale of the folder structure is genzel1\OS_DONT_DELETE\UMAP_Basic_OS\number_of_rat\processed\behavioural_mode\trial\HDF_file
# put in the path to the HDF file
hdf5_file_path = r"E:\Donders\0_rem_test\processed\CN\post_trial4_2017-09-28_14-03-38\post_trial4_2017-09-28_14-03-38_REM_dict.h5"

# Extract the group name from the HDF5 file path
group_name = os.path.basename(os.path.dirname(hdf5_file_path))

# Open the HDF5 file in read mode
with h5py.File(hdf5_file_path, 'r') as hdf_file:
    # Access the group within the HDF5 file
    group_name = 'post_trial4_2017-09-28_14-03-38'
    if group_name in hdf_file:
        # Access the keys within the group
        rem_dict_keys = list(hdf_file[group_name].keys())

        # Iterate through each REM period
        for rem_key in rem_dict_keys:
            start_end = hdf_file[group_name][rem_key]['start-end'][:] #extracts start and end of each REM period
            wavelet_transform = hdf_file[group_name][rem_key]['wavelet_transform'][:] #extracts wavelet transform of respective REM state with amplitude output (square to get power!)
            imfs = hdf_file[group_name][rem_key]['IMFs'][:] #extracts IMFs
            imf_frequencies = hdf_file[group_name][rem_key]['IMF_Frequencies'][:] #extracts frequencies of IMFs
            instantaneous_phases = hdf_file[group_name][rem_key]['Instantaneous Phases'][:] #extracts instantaneous phases of IMFs
            instantaneous_frequencies = hdf_file[group_name][rem_key]['Instantaneous Frequencies'][:] #extracts instantaneous frequencies of IMFs
            instantaneous_amplitudes = hdf_file[group_name][rem_key]['Instantaneous Amplitudes'][:] #extracts instantaneous amplitude of IMFs
            rem_cycles = hdf_file[group_name][rem_key]['Cycles'][:] #extracts REM cycles

        # Now you can use or print these values/shapes as needed
            print(f"Start-End for {rem_key}: {start_end}")
            print(f"The shape of wavelet_transform for {rem_key} is: {np.shape(wavelet_transform)}")
            print(f"IMFs for {rem_key}: {imfs}, Shape: {np.shape(imfs)}") #remove {imfs} for less output
            print(f"IMF Frequencies for {rem_key}: {imf_frequencies}, Shape: {np.shape(imf_frequencies)}") #remove {imf_frequencies} for less output
            print(f"Instantaneous Phases for {rem_key}: {instantaneous_phases}, Shape: {np.shape(instantaneous_phases)}") #remove {instantaneous_phases} for less output
            print(f"Instantaneous Frequencies for {rem_key}: {instantaneous_frequencies}, Shape: {np.shape(instantaneous_frequencies)}") #remove {instantaneous_frequencies} for less output
            print(f"Instantaneous Amplitudes for {rem_key}: {instantaneous_amplitudes}, Shape: {np.shape(instantaneous_amplitudes)}") #remove {instantaneous_amplitudes} for less output
            print(f"Cycles for {rem_key}: {rem_cycles}, Shape: {np.shape(rem_cycles)}") #remove {rem_cycles} for less output

    else:
        print(f"Group '{group_name}' not found in the HDF5 file.")
        
    num_rem_periods = len(rem_dict_keys)
    print(f"Number of REM periods: {num_rem_periods}") #print the number of REM periods
```
