```python
import h5py
```

```python
# Open the HDF file
file_path = r'E:\Donders\1\processed\CN\post_trial5_2017-09-28_14-55-18\post_trial5_2017-09-28_14-55-18_REM_dict.h5'
with h5py.File(file_path, 'r') as hdf_file:
    # Access the key 'post_trial1_2017-09-28_11-30-59'
    main_key = 'post_trial5_2017-09-28_14-55-18'

    # Check if the main key exists
    if main_key in hdf_file:
        # Iterate over the keys within the main key
        for key in hdf_file[main_key].keys():
            # Check if the key starts with 'REM'
            if key.startswith('REM'):
                print(f"Subdictionary under key '{key}':")

                # Access the subdictionary
                sub_dict = hdf_file[main_key][key]

                # Iterate over the keys in the subdictionary
                for sub_key in sub_dict.keys():
                    print(f"  Subdictionary '{sub_key}' has shape: {sub_dict[sub_key].shape}")
    else:
        print(f"No '{main_key}' key found in the HDF file.")


```
